```python
collection_faq = db["faq_data"]

# faq_data 컬렉션에서 전체 데이터를 가져옴
faq_data = list(collection_faq.find())

# MongoDB의 ObjectId를 JSON으로 직렬화할 수 없으므로 변환
for doc in faq_data:
    doc["_id"] = str(doc["_id"])  # ObjectId를 문자열로 변환

faq_data
```
